```python
#No6
a, b, c = input("a: "), input("b: "), input("c: ")
if a <= 0 or b <= 0 or c <= 0:
    print("Khong phai 3 canh cua tam giac")
elif max(a, b, c) >= (sum([a, b, c]) - max(a, b, c)):
    print("Khong phai 3 canh cua tam giac")
else:
    p = (a + b + c) / 2
    print((p(p - a)(p - b)(p - c)) ** 0.5)
```

```python
#No8
name = input("Ten chu ho: ")
electricityUsed = - int(input("Chi so thang truoc: ")) + int(input("Chi so thang nay"))
if electricityUsed <= 50:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(electricityUsed * 1984 * 1.08):.0f}")
elif electricityUsed <= 100:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(((electricityUsed - 50) * 2050 + 50 * 1984) * 1.08):.0f}")
elif electricityUsed <= 200:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(((electricityUsed - 100) * 2380 + 50 * (1984 + 2050)) * 1.08):.0f}")
elif electricityUsed <= 300:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(((electricityUsed - 200) * 2998 + 50 * (1984 + 2050) + 100 * 2380) * 1.08):.0f}")
elif electricityUsed <= 400:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(((electricityUsed - 300) * 3350 + 50 * (1984 + 2050) + 100 * (2380 + 2998)) * 1.08):.0f}")
else:
    print(f"Ho va ten: {name}")
    print(f"Tien phai tra la: {(((electricityUsed - 400) * 3460 + 50 * (1984 + 2050) + 100 * (2380 + 2998 + 3350)) * 1.08):.0f}")
```
