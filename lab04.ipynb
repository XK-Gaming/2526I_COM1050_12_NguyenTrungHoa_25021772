```python
#W4A3
n = int(input("Enter number: "))
while n < 0:
    n = int(input("Try again: "))
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
factorial(n)
```

```python
#W4A10
n = int(input("Enter number"))
highestCountNumber = 1
highestCount = 1
for i in range(1, n + 1):
    count = 1
    currentNumber = i
    while currentNumber != 1:
        if currentNumber % 2 == 0:
            currentNumber /= 2
        else:
            currentNumber = currentNumber * 3 + 1
        count += 1
    if count > highestCount:
        highestCount = count
        highestCountNumber = i
print(highestCountNumber, highestCount)
```
