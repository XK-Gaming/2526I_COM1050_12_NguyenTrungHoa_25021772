```python
#W6A1
def filterDupes(arr):
    result = []
    for i in arr:
        if i not in result:
            result.append(i)
    return result
filterDupes(list(map(int, input().split())))
```

```python
#W6A2
def sumToIndex(arr):
    result = []
    for i in range(len(arr)):
        result.append(sum(arr[:i+1]))
    return result
sumToIndex(input().split())
```

```python
#W6A3
def arrShift(arr, k):
    return arr[k%len(arr):] + arr[:k%len(arr)]
arrShift(input().split(), int(input()))
```

```python
#W6A4
def arrToDict(arr):
    resultDict = dict()
    for i in range(len(arr)):
        arr[i] = arr[i].split(":")
        if arr[i][0] not in resultDict:
            resultDict[arr[i][0]] = [arr[i][1]]
        else:
            resultDict[arr[i][0]].append(arr[i][1])
    return resultDict
arrToDict(input().split())
```

```python
#W6A5
def countNumber(arr):
    numbersCount = {
        "positive" : 0,
        "zero" : 0,
        "negative" : 0
    }
    for i in arr:
        if int(i) > 0:
            numbersCount["positive"] += 1
        elif int(i) == 0:
            numbersCount["zero"] += 1
        else:
            numbersCount["negative"] += 1
    return numbersCount
countNumber(input().split())
```

```python
#W6A6
def sumOfNumbers(arr):
    return sum([int(i) for i in arr])
sumOfNumbers(input().split())
```

```python
#W6A7
def tupleInfo(tup):
    return tup[0], tup[-1], tup[::-1]
tupleInfo(tuple(input().split()))
```

```python
#W6A8
def wordCount(arr):
    words = dict()
    for i in arr:
        if i not in words:
            words[i] = 1
        else:
            words[i] += 1
    return words
wordCount(input().split())
```

```python
#W6A9
def combineDict(a, b):
    resultDict = dict()
    aKeys = list(a.keys())
    bKeys = list(b.keys())
    while aKeys and bKeys:
        currKey = aKeys.pop()
        if currKey in b:
            resultDict[currKey] = a[currKey] + b[currKey]
            bKeys.remove(currKey)
        else:
            resultDict[currKey] = a[currKey]
    while aKeys:
        currKey = aKeys.pop()
        resultDict[currKey] = a[currKey]
    while bKeys:
        currKey = bKeys.pop()
        resultDict[currKey] = b[currKey]
    return(dict(sorted(resultDict.items())))
combineDict(, {"b":2})
```

```python
#W6A10
def pairSum(arr, k):
    result = []
    for i in range(len(arr)):
        for o in arr[i + 1:]:
            if arr[i] + o == k:
                result.append((arr[i], o))
    return sorted(result)
pairSum(list(map(int, input().split())), int(input()))
```

```python
#W6A11
def parityFilter(tup):
    even = []
    odd = []
    for i in tup:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return tuple(even), tuple(odd)
a, b = parityFilter(tuple(map(int, input().split())))
print(a)
print(b)
```

```python
#W6A12
def maxFreq(arr):
    freqs = dict()
    for i in arr:
        if i not in freqs:
            freqs[i] = 1
        else:
            freqs[i] += 1
    result = sorted(freqs)[0]
    for i in sorted(freqs):
        if freqs[i] > freqs[result]:
            result = i
    return result
maxFreq(list(map(int, input().split())))
```

```python
#W6A13
def keyValueSwitch(dictionary):
    result = dict()
    for i in dictionary:
        result[dictionary[i]] = i
    return result
```

```python
#W6A14
def intersection(a, b):
    result = set()
    for i in a:
        if i in b:
            result.add(i)
    return list(result)
intersection(list(map(int, input().split())), list(map(int, input().split())))
```

```python
#W6A15
def valueFilter(dictionary, k):
    result = dict()
    for i in dictionary:
        if dictionary[i] > k:
            result[i] = dictionary[i]
    return result
```
