```python
#11
a, b, c = int(input("a: ")), int(input("b: ")), int(input("c: "))
if a <= 0 or b <= 0 or c <= 0:
    print("Không phải tam giác")
elif max(a, b, c) >= sum([a, b, c]) - max(a, b, c):
    print("Không phải tam giác")
elif max(a, b, c) == min(a, b, c):
    print("Tam giác đều")
elif (sum([a, b, c]) - max(a, b, c)) / 2 == min(a, b, c) or (sum([a, b, c]) - min(a, b, c)) / 2 == max(a, b, c):
    print("Tam giác cân")
else:
    print("Tam giác thường")
```

```python
#13
electricity = float(input("kWh: "))
while electricity < 0:
    print("Try again")
    electricity = float(input("kWh: "))
if electricity <= 50:
    print(electricity * 1500)
elif electricity <= 100:
    print((electricity - 50) * 2000 + 50 * 1500)
else:
    print((electricity - 100) * 3000 + 50 * (1500 + 2000))
```
