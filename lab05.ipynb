```python
#W5A3
n = int(input())
for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
        print(False)
        break
else:
    print(n > 1)
```

```python
#W5A4
n = int(input())
result = 0
for i in range(1, n):
    if n % i == 0:
        result += i
print(result == n)
```

```python
#W5A7
calculation = input().split()
if calculation[1] == "+":
    print(round(float(calculation[0]) + float(calculation[2]), 2))
elif calculation[1] == "-":
    print(round(float(calculation[0]) - float(calculation[2]), 2))
elif calculation[1] == "*":
    print(round(float(calculation[0]) * float(calculation[2]), 2))
else:
    print(round(float(calculation[0]) / float(calculation[2]), 2))
```

```python
#W5A8
a, b = map(int, input().split())
count = 0
while a != 0 and b != 0:
    if a % 2 != b % 2:
        count += 1
    a //= 2
    b //= 2
while a != 0:
    if a % 2 == 1:
        count += 1
    a //= 2
while b != 0:
    if b % 2 == 1:
        count += 1
    b //= 2
print(count)
```

```python
#W5A10
a, b = map(str, input().split())
moves = dict()
for i in range(len(a)):
    if a[i] not in moves:
        moves[a[i]] = b[i]
    else:
        if moves[a[i]] != b[i]:
            print(False)
            break
else:
    if len(moves.values()) != len(set(moves.values())):
        print(False)
    else:
        print(True)
```
